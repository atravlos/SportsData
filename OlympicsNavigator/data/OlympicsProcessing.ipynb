```python
import pandas as pd

# Load csv file as dataframe
df = pd.read_csv('athlete_events.csv')
# Drop unwanted columns
df = df.drop(['Sex', 'Age', 'Height', 'Weight', 'Team', 'Games', 'City'], axis=1)
# Include only medalists
df = df[df["Medal"].isin(["Gold", "Silver", "Bronze"])]
# Convert Year to String (so commas aren't seen in the years)
df["Year"] = df["Year"].astype(str)

df.head(10)
```

```python
df.to_csv('olympicmedalists.csv', index=False)
```
